```python
import cv2

def main():
    cap = cv2.VideoCapture(0)  # Use default webcam (index 0)
    
    # Check if the webcam opened successfully
    if not cap.isOpened():
        print("Error: Could not open webcam.")
        return
    
    # Read the initial frame
    ret, frame1 = cap.read()
    if not ret:
        print("Error: Could not read frame from webcam.")
        return
    
    # Convert frame1 to grayscale
    gray1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
    
    while True:
        # Read another frame from the webcam
        ret, frame2 = cap.read()
        if not ret:
            print("Error: Could not read frame from webcam.")
            break
        
        # Convert frame2 to grayscale
        gray2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
        
        # Compute the absolute difference between frame1 and frame2
        diff = cv2.absdiff(gray1, gray2)
        
        # Apply a threshold to create a binary image
        _, thresh = cv2.threshold(diff, 30, 255, cv2.THRESH_BINARY)
        
        # Dilate the thresholded image to fill in holes
        dilated = cv2.dilate(thresh, None, iterations=2)
        
        # Find contours of the dilated image
        contours, _ = cv2.findContours(dilated, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        
        # Draw rectangles around the detected motion
        for contour in contours:
            (x, y, w, h) = cv2.boundingRect(contour)
            if cv2.contourArea(contour) < 700:  # Adjust this threshold as needed
                continue
            cv2.rectangle(frame1, (x, y), (x+w, y+h), (0, 255, 0), 2)
        
        # Display the original frame with motion rectangles
        cv2.imshow('Motion Detection', frame1)
        
        # Update frame1 to be frame2 for the next iteration
        frame1 = frame2
        gray1 = gray2
        
        # Check for user input to quit the program
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    # Release the webcam and close all OpenCV windows
    cap.release()
    cv2.destroyAllWindows()

if _name_ == "_main_":
    main()
```
